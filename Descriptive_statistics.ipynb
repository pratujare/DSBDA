```python
def display_statistics(dataset, dataset_name):
    numerical_columns = ['age', 'fnlwgt', 'educational-num', 'capital-gain', 'capital-loss', 'hours-per-week']
    print(f"\nStatistics for {dataset_name}:")

    # Median calculation without using built-in median
    def median(column_data):
        sorted_data = sorted(column_data)
        n = len(sorted_data)
        mid = n // 2
        if n % 2 == 0:
            return (sorted_data[mid - 1] + sorted_data[mid]) / 2
        return sorted_data[mid]

    # Min and max using built-in functions
    def min_max(column_data):
        return min(column_data), max(column_data)

    # MEAN
    print('\nMean')
    for col in numerical_columns:
        values = dataset[col].dropna().values
        mean = sum(values) / len(values)  # Manual mean
        print(f"{col}: {mean:.3f}")

    # STANDARD DEVIATION
    print('\nStandard Deviation')
    for col in numerical_columns:
        values = dataset[col].dropna().values
        mean = sum(values) / len(values)
        variance = sum((x - mean) ** 2 for x in values) / len(values)
        std_dev = variance ** 0.5
        print(f"{col}: {std_dev:.6f}")

    # MEDIAN
    print("\nMedian")
    for col in numerical_columns:
        median_value = median(dataset[col].dropna().values)
        print(f"{col}: {median_value:.3f}")

    # MIN and MAX
    print("\nMin and Max")
    for col in numerical_columns:
        min_value, max_value = min_max(dataset[col].dropna().values)
        print(f"{col}: Min = {min_value}, Max = {max_value}")

    # 25th Percentile
    print("\nQuantile 25")
    for col in numerical_columns:
        q1 = dataset[col].quantile(0.25)
        print(f"25th Percentile of {col}: {q1}")

    # 75th Percentile
    print("\nQuantile 75")
    for col in numerical_columns:
        q3 = dataset[col].quantile(0.75)
        print(f"75th Percentile of {col}: {q3}")

# Example usage:
display_statistics(df, "Demographic Dataset")
```

```python
def display_statistics(dataset, dataset_name):
    numerical_columns = ['age', 'fnlwgt', 'educational-num', 'capital-gain', 'capital-loss', 'hours-per-week']
    print(f"\nStatistics for {dataset_name}:")

    print("\nMean")
    for col in numerical_columns:
        values = list(dataset[col])
        total = sum(values)
        count = len(values)
        mean = total / count
        print(f"{col}: {mean:.3f}")

    print("\nStandard Deviation")
    for col in numerical_columns:
        values = list(dataset[col])
        total = sum(values)
        count = len(values)
        mean = total / count
        variance = sum((x - mean) ** 2 for x in values) / count
        std_dev = variance ** 0.5
        print(f"{col}: {std_dev:.3f}")

    print("\nMedian")
    for col in numerical_columns:
        values = sorted(list(dataset[col]))
        n = len(values)
        if n % 2 == 0:
            median = (values[n//2 - 1] + values[n//2]) / 2
        else:
            median = values[n//2]
        print(f"{col}: {median:.3f}")

    print("\nMin and Max")
    for col in numerical_columns:
        values = list(dataset[col])
        minimum = min(values)
        maximum = max(values)
        print(f"{col}: Min = {minimum}, Max = {maximum}")

    print("\n25th Percentile (Q1)")
    for col in numerical_columns:
        values = sorted(list(dataset[col]))
        index = int(0.25 * len(values))
        q1 = values[index]
        print(f"{col}: {q1}")

    print("\n75th Percentile (Q3)")
    for col in numerical_columns:
        values = sorted(list(dataset[col]))
        index = int(0.75 * len(values))
        q3 = values[index]
        print(f"{col}: {q3}")
        
display_statistics(df,"Demographic Dataset")
```

```python
import numpy as np

def display_statistics(species_data, species_name):
    nc = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
    print(f"\nStatistics for {species_name}:")

    # Mean
    print("\nMean")
    for col in nc:
        column_data = species_data[col].values
        mean = sum(column_data) / len(column_data)
        print(f"{col}    {mean:.3f}")

    # Standard Deviation
    print("\nStandard Deviation")
    for col in nc:
        column_data = species_data[col].values
        mean = sum(column_data) / len(column_data)
        variance = sum((x - mean) ** 2 for x in column_data) / len(column_data)
        std_deviation = np.sqrt(variance)
        print(f"{col}    {std_deviation:.6f}")

    # Min and Max
    print("\nMin and Max")
    for col in nc:
        column_data = species_data[col].values
        min_value = min(column_data)
        max_value = max(column_data)
        print(f"{col}    Min: {min_value:.3f}    Max: {max_value:.3f}")

    # Median
    print("\nMedian")
    for col in nc:
        column_data = sorted(species_data[col].values)
        n = len(column_data)
        if n % 2 == 0:  # Even number of elements
            median = (column_data[n//2 - 1] + column_data[n//2]) / 2
        else:  # Odd number of elements
            median = column_data[n//2]
        print(f"{col}    {median:.3f}")

    # Quantile 25
    print("\nQuantile 25")
    for col in nc:
        column_data = sorted(species_data[col].values)
        q1 = np.percentile(column_data, 25)
        print(f"25th Percentile of {col}: {q1:.3f}")

    # Quantile 75
    print("\nQuantile 75")
    for col in nc:
        column_data = sorted(species_data[col].values)
        q3 = np.percentile(column_data, 75)
        print(f"75th Percentile of {col}: {q3:.3f}")

# Example usage:
display_statistics(setosa_data, 'Setosa')
```

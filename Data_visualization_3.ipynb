```python
# Plot histogram for each numeric feature
df.hist(figsize=(10, 8), bins=10)
plt.suptitle("Histograms of Iris Features", fontsize=16)
plt.show()
```

```python
# Boxplots for numeric features only
df.drop(columns='species').plot(kind='box', subplots=True, layout=(2,2), figsize=(10, 8), sharex=False)
plt.suptitle("Boxplots of Iris Features", fontsize=16)
plt.show()
```

```python
# Use seaborn to create boxplots
for column in df.columns[:-1]:  # exclude 'species'
    sns.boxplot(y=df[column])
    plt.title(f"Boxplot of {column}")
    plt.show()
```

```python
plt.subplot(1,2,1)
sns.histplot(df["sepal_length"], bins=20, kde=True)
plt.title("Distribution of Sepal length")
plt.xlabel("Sepal length (cm) ")
plt.ylabel("count")

plt.subplot(1,2,2)
sns.histplot(df["petal_length"], bins=20, kde=True)
plt.title("Distribution of Petall length")
plt.xlabel("Petal length (cm) ")
plt.ylabel("count")

plt.tight_layout()
plt.show()
```

```python
#The maximum sepal width is 3.0
sns.histplot(x='sepal_width', data=df)
```

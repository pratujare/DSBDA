```python
numerical_columns=df.select_dtypes(include=[np.number]).columns
skewness=df[numerical_columns].apply(lambda x: stats.skew(x,nan_policy='omit'))
print("\nSkewness of numerical columns: ")
print(skewness)
```

```python
high_skew_columns=skewness[skewness.abs()>1].index
for column in high_skew_columns:
    plt.figure(figsize=(10,5))
    sns.histplot(df[column],kde=True,bins=30)
    plt.title(f'Distribution of {column} (Skewness: {skewness[column]:.2f})')
    plt.show()
```

```python
print(df['Attendance'].isnull().sum())
print(np.isfinite(df['Attendance']).all())
print(df['Attendance'].head())
```

```python
for column in high_skew_columns:
    if skewness[column] > 1:
        if (df[column] > 0).all():
            df[column] = np.log(df[column])
            print(f"\nApplied log transformation to {column}")
        else:
            print(f"\nSkipped log transformation for {column} due to non-positive values")

    elif skewness[column] < -1:
        if (df[column] >= 0).all():
            df[column] = np.sqrt(df[column])
            print(f"\nApplied square root transformation to {column}")
        else:
            print(f"\nSkipped square root transformation for {column} due to negative values")

if skewness['Attendance'] > 1 or skewness['Attendance'] < -1:
    if (df['Attendance'] > 0).all():
        df['Attendance'], _ = stats.boxcox(df['Attendance'])
        print("\nApplied Box-Cox transformation to 'Attendance' for skewness reduction")
    else:
        min_value = df['Attendance'].min()
        shift_value = abs(min_value) + 1
        df['Attendance'] = df['Attendance'] + shift_value
        print(f"\nShifted 'Attendance' by {shift_value} to make all values positive")
        
        df['Attendance'], _ = stats.boxcox(df['Attendance'])
        print("\nApplied Box-Cox transformation to 'Attendance' after shifting values")
```
